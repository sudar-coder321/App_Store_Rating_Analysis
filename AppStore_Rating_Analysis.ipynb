```python
li2=[]
li3=[]
i=0
for x in li1:
    li2=x.split(', ')
    for y in li2:
        li3.append([y,li[i]])
    i+=1
```

```python
li4=[]
sum1=0
flg=0
for i in range(len(li3)):
    sum1=li3[i][1]
    flg=0
    for k in range(i-1,-1,-1):
        if(li3[k][0]==li3[i][0]):
            flg=1
            break
    if(flg==1):
        continue
    for j in range(i+1,len(li3)):
        if(li3[j][0]==li3[i][0]):
            sum1+=li3[j][1]
    li4.append([li3[i][0],int(sum1)])
```

```python
genrelist=[]
userratlist=[]
for i in range(len(li4)):
    if(li4[i][0]!='Games' and li4[i][0]!='Strategy'):
        genrelist.append(li4[i][0])
        userratlist.append(li4[i][1])
```

```python
li4.sort(key=lambda x: x[1])
li4.reverse()
print("The genres with their ratings are:-")
for i in range(len(li4)):
    print(i,end=") ")
    print(li4[i])
```

```python
noofgen=len(li4)
print("Enter the number of genres to be taken as one combination to determine the best combination possible(as input below)")
noofgensel=int(input())
```

```python
print("The best combination of genres formed is")
for i in range(noofgensel):
    print(li4[i])
```

```python
mpt.figure(1,figsize=(20,50),dpi=80, facecolor='w', edgecolor='k')    
mpt.barh(genrelist,userratlist,height=0.5,left=None,align='center',color=['orange'])
mpt.ylabel('Genres')
mpt.xlabel('No. of User Ratings above 4.0')
mpt.title('Analysis of Genres which have user rating above 4.0')

```
